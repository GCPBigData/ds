```python
import seaborn as sns
sns.set()
```

```python
import pandas as pd
athletes = pd.read_csv('athletes.csv')
athletes = athletes[athletes['sport'].isin(['wrestling', 'badminton', 'triathlon', 'gymnastics'])]
```

```python
athletes.head()
```

# Relational Plots

```python
sns.relplot(data=athletes, x='weight', y='height', col='sport', col_wrap=2, hue='sex',
           kind='line')
```

```python
df = pd.read_csv('weather.csv') 
aug = df[df['MONTH'] == 8]
aug.head()
```

```python
sns.relplot(data=aug, x='TIME', y='TEMP', kind='line')
```

# Categorical Plots

```python
sns.catplot(data=athletes, x="sport", y="weight", hue='sex', kind='boxen');
```
