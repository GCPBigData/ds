## 4.2 Replacing Bad Data with NaN

```python
import pandas as pd
```

```python
data = pd.read_csv('_datasets/artwork_data.csv', low_memory=False)
```

```python
data.head()
```

```python
pd.isna(data.loc[:, 'dateText'])
```

```python
from numpy import nan
```

```python
data.replace({ 'dateText': { 'date not known': nan }})
```

```python
data.head()
```

```python
data.replace({ 'dateText': { 'date not known': nan }}, inplace=True)
```

```python
data.head()
```

```python
data = pd.read_csv('_datasets/artwork_data.csv', low_memory=False)
```

```python
data.head()
```

```python
data.loc[data.dateText == 'date not known', ['dateText']] = nan
```

```python
data.head()
```

```python
data.loc[data.year.notnull() & data.year.astype(str).str.contains('[^0-9]')]
```

```python
data.loc[data.year.notnull() & data.year.astype(str).str.contains('[^0-9]'), ['year']] = nan
```
